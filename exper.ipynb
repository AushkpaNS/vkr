```python
import sys
import os
import json
import matplotlib.pyplot as plt
import numpy as np

sys.path.insert(0, os.path.abspath('..'))

from main import item_knn

DEFAULT_GRIDS = {
    "50m": {
        "likes": [0, 0.001, 0.25, 0.5, 1, 2, 3, 4, 6],
        "listens": [0, 0.001, 0.002, 0.004, 0.008, 0.016, 0.032, 0.064, 0.128, 0.256, 0.5, 1.0, 2],
    },
    "500m": {
        "likes": [0, 0.002, 0.004, 0.008, 0.016, 0.032, 0.064, 0.128, 0.256, 0.5, 1.0],
        "listens": [0, 0.001, 0.002, 0.004, 0.008],
    },
}

base_params = {
    'data_dir': "../../data/flat",
    'size': "50m",
    'interaction': "likes",
    'device': "cpu",
    'hours': DEFAULT_GRIDS["50m"]["likes"],
    'validation_metric': "ndcg@100",
    'report_metrics': ["ndcg@10", "ndcg@50", "ndcg@100", 
                      "dcg@10", "dcg@50", "dcg@100",
                      "recall@10", "recall@50", "recall@100",
                      "coverage@10", "coverage@50", "coverage@100"]
}

# генерируем значения deep от 864000 до 26000000 с шагом 864000
start_deep = 864000
end_deep = 26000000
step = 864000

deep_values = list(range(start_deep, end_deep, step))
deep_values.append(end_deep)  # добавляем последнее значение 26000000

print(f"Запуск экспериментов для {len(deep_values)} значений deep:")
print(f"Значения deep: {deep_values}")

# словарь для хранения результатов всех экспериментов
all_results = []

# запускаем эксперименты для каждого значения deep
for deep_val in deep_values:
    print(f"\nЗапуск эксперимента с deep={deep_val:,}...")
    
    current_params = base_params.copy()
    current_params['deep'] = deep_val
    
    try:
        result = item_knn(**current_params)
        result['deep'] = deep_val
        all_results.append(result)
        print(f"Успешно завершено для deep={deep_val:,}")
        
        print(f"  ndcg@100: {result['ndcg'][100]:.4f}")
        print(f"  recall@100: {result['recall'][100]:.4f}")
        print(f"  coverage@100: {result['coverage'][100]:.4f}")
        
    except Exception as e:
        print(f"Ошибка при deep={deep_val:,}: {e}")
        all_results.append({
            'deep': deep_val,
            'error': str(e),
            'ndcg': {10: 0, 50: 0, 100: 0},
            'dcg': {10: 0, 50: 0, 100: 0},
            'recall': {10: 0, 50: 0, 100: 0},
            'coverage': {10: 0, 50: 0, 100: 0}
        })

output_file = "deep_experiments_results.json"
with open(output_file, 'w') as f:
    json.dump(all_results, f, indent=2)
print(f"\nВсе результаты сохранены в файл: {output_file}")

deep_vals = [r['deep'] for r in all_results]

metrics_data = {
    'ndcg@10': [],
    'ndcg@50': [],
    'ndcg@100': [],
    'dcg@10': [],
    'dcg@50': [],
    'dcg@100': [],
    'recall@10': [],
    'recall@50': [],
    'recall@100': [],
    'coverage@10': [],
    'coverage@50': [],
    'coverage@100': []
}

for result in all_results:
    metrics_data['ndcg@10'].append(result['ndcg'][10])
    metrics_data['ndcg@50'].append(result['ndcg'][50])
    metrics_data['ndcg@100'].append(result['ndcg'][100])
    
    metrics_data['dcg@10'].append(result['dcg'][10])
    metrics_data['dcg@50'].append(result['dcg'][50])
    metrics_data['dcg@100'].append(result['dcg'][100])
    
    metrics_data['recall@10'].append(result['recall'][10])
    metrics_data['recall@50'].append(result['recall'][50])
    metrics_data['recall@100'].append(result['recall'][100])
    
    metrics_data['coverage@10'].append(result['coverage'][10])
    metrics_data['coverage@50'].append(result['coverage'][50])
    metrics_data['coverage@100'].append(result['coverage'][100])

def format_deep_value(val):
    if val >= 1_000_000:
        return f"{val/1_000_000:.0f}M"
    elif val >= 1000:
        return f"{val/1000:.0f}K"
    return str(val)

fig, axes = plt.subplots(4, 3, figsize=(18, 15))
fig.suptitle('Зависимость метрик от значения deep (864K до 26M)', fontsize=16, fontweight='bold')

plot_configs = [
    (axes[0, 0], 'ndcg@10', 'NDCG@10', 'blue'),
    (axes[0, 1], 'ndcg@50', 'NDCG@50', 'blue'),
    (axes[0, 2], 'ndcg@100', 'NDCG@100', 'blue'),
    (axes[1, 0], 'dcg@10', 'DCG@10', 'green'),
    (axes[1, 1], 'dcg@50', 'DCG@50', 'green'),
    (axes[1, 2], 'dcg@100', 'DCG@100', 'green'),
    (axes[2, 0], 'recall@10', 'Recall@10', 'red'),
    (axes[2, 1], 'recall@50', 'Recall@50', 'red'),
    (axes[2, 2], 'recall@100', 'Recall@100', 'red'),
    (axes[3, 0], 'coverage@10', 'Coverage@10', 'purple'),
    (axes[3, 1], 'coverage@50', 'Coverage@50', 'purple'),
    (axes[3, 2], 'coverage@100', 'Coverage@100', 'purple'),
]

for ax, metric_key, title, color in plot_configs:
    ax.plot(deep_vals, metrics_data[metric_key], 'o-', color=color, linewidth=2, markersize=4)
    ax.set_title(title, fontsize=12, fontweight='bold')
    ax.set_xlabel('deep value')
    ax.set_ylabel('Metric Value')
    ax.grid(True, alpha=0.3)
    
    ax.set_xticks([deep_vals[0], deep_vals[len(deep_vals)//2], deep_vals[-1]])
    ax.set_xticklabels([format_deep_value(deep_vals[0]), 
                       format_deep_value(deep_vals[len(deep_vals)//2]), 
                       format_deep_value(deep_vals[-1])])

plt.tight_layout()
plt.savefig('deep_experiments_metrics.png', dpi=150, bbox_inches='tight')
print(f"\nГрафики сохранены в файл: deep_experiments_metrics.png")

fig2, ax2 = plt.subplots(2, 2, figsize=(14, 10))
fig2.suptitle('Сводная зависимость ключевых метрик от deep', fontsize=16, fontweight='bold')

summary_metrics = [
    (ax2[0, 0], 'ndcg@100', 'NDCG@100', 'blue'),
    (ax2[0, 1], 'recall@100', 'Recall@100', 'red'),
    (ax2[1, 0], 'coverage@100', 'Coverage@100', 'green'),
]

for i, (ax, metric_key, title, color) in enumerate(summary_metrics):
    ax.plot(deep_vals, metrics_data[metric_key], 'o-', color=color, linewidth=3, markersize=6)
    ax.set_title(title, fontsize=14, fontweight='bold')
    ax.set_xlabel('Значение deep', fontsize=11)
    ax.set_ylabel('Значение метрики', fontsize=11)
    ax.grid(True, alpha=0.3)
    
    max_val = max(metrics_data[metric_key])
    max_idx = metrics_data[metric_key].index(max_val)
    ax.annotate(f'Max: {max_val:.4f}\n@deep={format_deep_value(deep_vals[max_idx])}', 
                xy=(deep_vals[max_idx], max_val),
                xytext=(10, 10), textcoords='offset points',
                bbox=dict(boxstyle="round,pad=0.3", facecolor="yellow", alpha=0.7))

fig2.delaxes(ax2[1, 1])

plt.tight_layout()
plt.savefig('deep_experiments_summary.png', dpi=150, bbox_inches='tight')
print(f"Сводный график сохранен в файл: deep_experiments_summary.png")

print("\n" + "="*80)
print("СВОДНАЯ ТАБЛИЦА РЕЗУЛЬТАТОВ:")
print("="*80)

print("\nЛучшие значения метрик:")
for metric in ['ndcg@100', 'recall@100', 'coverage@100']:
    values = metrics_data[metric]
    best_idx = values.index(max(values))
    print(f"{metric:15} = {values[best_idx]:.6f} (при deep={format_deep_value(deep_vals[best_idx])})")

print("\n" + "="*80)
print(f"Всего проведено экспериментов: {len(all_results)}")
print(f"Диапазон deep: от {format_deep_value(deep_vals[0])} до {format_deep_value(deep_vals[-1])}")

plt.show()
```
